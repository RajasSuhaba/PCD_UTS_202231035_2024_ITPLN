## Ambang Batas Red_Blue

```python
def display_image(image, title="Image"):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('on')
    plt.show()

# Tentukan rentang warna merah dalam HSV
lower_red = np.array([0, 50, 50])         # Ambang bawah warna merah (dalam HSV)
upper_red = np.array([10, 255, 255])      # Ambang atas warna merah (dalam HSV)
lower_red2 = np.array([170, 50, 50])      # Ambang bawah tambahan warna merah (dalam HSV)
upper_red2 = np.array([180, 255, 255])    # Ambang atas tambahan warna merah (dalam HSV)

# Tentukan rentang warna biru dalam HSV
lower_blue = np.array([100, 50, 50])      # Ambang bawah warna biru (dalam HSV)
upper_blue = np.array([130, 255, 255])    # Ambang atas warna biru (dalam HSV)

# Buat mask untuk menangkap piksel merah dalam rentang tersebut
red_mask1 = cv2.inRange(hsv_image, lower_red, upper_red)
red_mask2 = cv2.inRange(hsv_image, lower_red2, upper_red2)

# Buat mask untuk menangkap piksel biru dalam rentang tersebut
blue_mask = cv2.inRange(hsv_image, lower_blue, upper_blue)

# Gabungkan kedua mask untuk menangkap piksel yang berwarna merah atau biru
red_blue_mask = cv2.bitwise_or(red_mask1, cv2.bitwise_or(red_mask2, blue_mask))

# Buat citra putih dengan ukuran yang sama dengan citra asli
white_image = np.full(image.shape, 255, dtype=np.uint8)
red_blue_detected_image = cv2.bitwise_and(white_image, white_image, mask=red_blue_mask)
display_image(red_blue_detected_image, "Red-Blue")
```

## Ambang Batas Red-Green-Blue

```python
def display_image(image, title="Image"):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('on')
    plt.show()

# Tentukan rentang warna merah dalam HSV
lower_red1 = np.array([0, 50, 50])        # Ambang bawah warna merah (dalam HSV)
upper_red1 = np.array([10, 255, 255])     # Ambang atas warna merah (dalam HSV)
lower_red2 = np.array([170, 50, 50])      # Ambang bawah tambahan warna merah (dalam HSV)
upper_red2 = np.array([180, 255, 255])    # Ambang atas tambahan warna merah (dalam HSV)

# Tentukan rentang warna hijau dalam HSV
lower_green = np.array([40, 50, 50])      # Ambang bawah warna hijau (dalam HSV)
upper_green = np.array([80, 255, 255])    # Ambang atas warna hijau (dalam HSV)

# Tentukan rentang warna biru dalam HSV
lower_blue = np.array([100, 50, 50])      # Ambang bawah warna biru (dalam HSV)
upper_blue = np.array([130, 255, 255])    # Ambang atas warna biru (dalam HSV)

# Buat mask untuk menangkap piksel merah, hijau, dan biru dalam rentang tersebut
red_mask1 = cv2.inRange(hsv_image, lower_red1, upper_red1)
red_mask2 = cv2.inRange(hsv_image, lower_red2, upper_red2)
green_mask = cv2.inRange(hsv_image, lower_green, upper_green)
blue_mask = cv2.inRange(hsv_image, lower_blue, upper_blue)

# Gabungkan ketiga mask untuk menangkap piksel yang berwarna merah, hijau, dan biru
red_green_blue_mask = cv2.bitwise_or(red_mask1, cv2.bitwise_or(red_mask2, cv2.bitwise_or(green_mask, blue_mask)))

# Buat citra putih dengan ukuran yang sama dengan citra asli
white_image = np.full(image.shape, 255, dtype=np.uint8)
red_green_blue_detected_image = cv2.bitwise_and(white_image, white_image, mask=red_green_blue_mask)
display_image(red_green_blue_detected_image, "Red-Green-Blue")
```

## Ambang Batas None
